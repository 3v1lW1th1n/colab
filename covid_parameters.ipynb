```
mortality_rate = 0.01
doubling_time = 5
time_infect_to_death = 20
```

```
# estimate infections at 1st death
# สมมติว่า วันนี้ เจอคนตายคนแรก แปลว่า มีคนติดเชื้อแล้วกี่คน
n0 = 1/mortality_rate
print(f'{n0} people first infected, {time_infect_to_death} days ago.')
dub = time_infect_to_death / doubling_time
print(f'Then to now, {dub} doublings.')
N = n0 * 2**dub
print(f'So, today has {N} infected people.')
```
